```python
import json
from langdetect import detect

# Your original input
original_data = {
    "1": "You are not that stupid.",
    "2": "You are stupid",
    "3": "Schiesse! du bist dumm",
    "4": "I hate nothing!",
    "5": "algo esta mal",
    "6": "No eres tan inocente",
    "7": "You're such a fool.",
    "8": "That was a dumb idea.",
    "9": "Shut your mouth.",
    "10": "This is the worst thing you've ever said.",
    "11": "Est치s loco.",
    "12": "Ni siquiera puedes pensar.",
    "13": "You can't do anything right.",
    "14": "Nobody likes you.",
    "15": "How can someone be this clueless?",
    "16": "That was pathetic.",
    "17": "Stop being so annoying.",
    "18": "Du bist so nutzlos.",
    "19": "C치llate ya.",
    "20": "You really messed this up."
}
# with open('cleaned_data.json', 'r')as f:
#     original_data = json.load(f)


# Basic placeholder sentiment analysis
def assign_sentiment(text):
    if any(word in text.lower() for word in ["stupid", "fool", "dumb", "shut", "worst", "clueless", "pathetic", "annoying", "useless", "mess", "loco"]):
        return "Negative"
    elif "not" in text.lower():
        return "Positive"
    else:
        return "Neutral"

# Detect language and enrich the data
converted_data = {
    k: {
        "TEXT": v,
        "Sentiment": assign_sentiment(v),
        "Language": detect(v).upper()
    } for k, v in original_data.items()
}

# Output the result
print(json.dumps(converted_data, indent=4, ensure_ascii=False))
```
